```python
random.seed(4)
num_of_projects = 1000
projects = []

for i in range(num_of_projects):

    # create project_id
    project_id = i + 1

    # create team_size
    team_size = random.randint(4,10)

    # create budget ($)
    budget = random.randint(1,100) * 1000

    # create workload
    workload = random.choice(["S", "M", "L"])

    # create completion_time (days)
    completion_time = random.randint(1,10)

    # add info to list of projects
    projects.append([project_id, team_size, budget, workload, completion_time])

projects_df = pd.DataFrame(projects, columns = ["project_id", "team_size", "budget", "workload", "completion_time"])
```

```python
random.seed(4)
num_of_tasks = 4000
tasks = []
date_1, date_2 = date(2024, 2, 13), date(2024, 2, 29)
interval = date_2 - date_1
interval_days = interval.days

for i in range(num_of_tasks):

    # create task_id
    task_id = i + 1

    # create project_id
    project_id = random.randint(1, 1000)

    # create user_id
    user_id = random.randint(1, 500)

    # create status
    status = random.choice(["complete", "incomplete"])

    # create estimated_duration
    estimated_duration = random.randint(1,8)

    # create due_date
    due_date = str(date_1 + timedelta(days = random.randint(1,interval_days)))

    # create description
    description = random.choice(descriptions)

    # add info to list of tasks
    tasks.append([task_id, project_id, user_id, status, estimated_duration, due_date, description])

tasks_df = pd.DataFrame(tasks, columns = ["task_id", "project_id", "user_id", "status", "estimated_duration", "due_date", "description"])
```

```python
random.seed(4)
num_of_users = 500
users = []

for i in range(num_of_users):

    # create user_id
    user_id = i + 1

    # create username
    username = names.get_first_name() + str(user_id)

    # create password
    password = "".join(random.choices(string.ascii_lowercase + string.digits, k=4))

    # create role
    role = random.choices(["manager", "member"], weights = [20, 80], k=1)[0]

    # add info to list of tasks
    users.append([user_id, username, password, role])

users_df = pd.DataFrame(users, columns = ["user_id", "username", "password", "role"])
```
